```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame(np.random.randn(5,3),
                 columns = ['C1','C2','C3'])
df
```

```python
# ix : index로 indexing
df.ix[0,0] = None
df.ix[1,['C1','C3']] = np.nan
df.ix[2,'C2'] = np.nan
df.ix[3,'C2'] = np.nan
df.ix[4,'C3'] = np.nan

df
```

## 1. df.fillna()
옵션
- method='ffill' (='pad'), 'bfill' (=backfill) : 위/아래 row값으로 결측치 대체
- limit = 숫자 : 위/아래 row로 결측값 대체하는 회수 제한

```python
df.fillna(0)
```

```python
df.fillna('missing')
```

```python
# method = 'ffill' : 윗 row 값으로 null 채움
df.fillna(method='ffill')
```

```python
# method = 'bfill' : 아래 row 값으로 null 채움
df.fillna(method='bfill')
```

```python
# 대체 limit 설정
df.fillna(method='bfill', limit = 1)
```

```python
# 평균값으로 대체
df.fillna(df.mean())
```

```python
# same result
df.where(pd.notnull(df), df.mean(), axis = 'columns')
```

```python
# C1, C2컬럼만 평균값으로 대체
df.fillna(df.mean()['C1':'C2'])
```

```python
# np.where 이용 : np.where(조건, T, F)
df2 = pd.DataFrame({
    'C1' :[1,2,3,4,5],
    'C2' : [6,7,8,9,10]
})
df2
```

```python
df2.ix[[1,3],['C2']] = np.nan
df2
```

```python
# c2 컬럼값이 null 이면 c1컬럼값으로, notnull이면 그대로
df2['C2_new'] = np.where(pd.notnull(df2['C2']), df2['C2'], df2['C1'])
df2
```

## 2. dropna()
옵션
- axis = 0 : 행 전체 삭제
- axis = 1 : 열 전체 삭제

```python
df = pd.DataFrame(np.random.randn(5,4),
                 columns = ['C1','C2','C3','C4'])

df.ix[[0,1],'C1'] = None
df.ix[2,'C2'] = np.nan

df
```

```python
# 행 전체 삭제
df.dropna(axis =0)
```

```python
# 열 전체 삭제
df.dropna(axis = 1)
```

```python
df
```

```python
# C1 컬럼에서 결측치가 있는 행 삭제
df['C1'].dropna()
```

```python
df[['C1','C2','C3']].dropna(axis = 1)
```

```python
df.ix[[2,4],['C1','C2','C3']].dropna(axis = 1)
```
